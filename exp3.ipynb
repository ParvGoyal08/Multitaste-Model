```python
df.drop(columns=[ "Source"], axis=1, inplace=True)
```

```python
import pandas as pd
train_df = pd.read_csv("umami_training_dataset.csv")
test_df = pd.read_csv("Umami_testing_dataset.csv")

# Concatenate the DataFrames
umami_df = pd.concat([train_df, test_df], ignore_index=True)
```

```python
umami_df.columns
```

```python
umami_df.drop(columns=["Column1", "SEQUENCE", "Parent_SMILES"], inplace=True)
```

```python
display(umami_df.head())
```

```python
umami_df.rename(columns={"TASTE": "Canonicalized Taste", "Std_SMILES": "Standardized SMILES"}, inplace=True)
```

```python
umami_df = umami_df[umami_df["Canonicalized Taste"] == "umami"].reset_index(drop=True)
```

```python
combined_df = pd.concat([df, umami_df], ignore_index=True)  
```

```python
display(combined_df.head())
```

```python
combined_df["Canonicalized Taste"].value_counts()
```

```python
combined_df.describe()
```

```python
combined_df = combined_df.drop_duplicates(subset="Standardized SMILES").reset_index(drop=True)
combined_df['Canonicalized Taste'].value_counts()
```

```python
from rdkit import Chem

# Function to canonicalize a SMILES string
def canonicalize_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        return Chem.MolToSmiles(mol, canonical=True)
    else:
        return None

# Apply the function to the 'Standardized SMILES' column
combined_df['Canonicalized SMILES'] = combined_df['Standardized SMILES'].apply(canonicalize_smiles)
```

```python
combined_df.to_csv("combined_deduplicated.csv", index=False)
```
